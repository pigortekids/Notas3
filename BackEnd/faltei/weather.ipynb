```
Y = df_na_num.RainTomorrow
X = df_na_num.drop(['RainTomorrow'], axis = 1)
```

```
from sklearn import preprocessing # To get MinMax Scaler function

#Normalizando a entrada
min_max_scaler = preprocessing.MinMaxScaler()

X_norm = pd.DataFrame( min_max_scaler.fit_transform(X) , columns = X.columns)


from sklearn.metrics import accuracy_score # To calculate the score
from sklearn.feature_selection import SelectKBest # Univariate Feature Selection
from sklearn.feature_selection import chi2 # To apply Univariate Feature Selection
# Initialize SelectKBest function
UnivariateFeatureSelection = SelectKBest(chi2, k=13).fit(X_norm, Y)



# Cria a lista para exibição
diccionario = {key:value for (key, value) in zip(UnivariateFeatureSelection.scores_, X_norm.columns)}
sorted(diccionario.items())

```

```
X_top4 = X_norm[['RainToday', 'Humidity3pm','Rainfall','Humidity9am']]
X_top4.head()
```

```
from sklearn.model_selection import train_test_split
x_train,x_test_val,y_train,y_test_val = train_test_split(X_top4,Y,test_size = 0.4,random_state = 40)
x_test, x_val, y_test, y_val = train_test_split(x_test_val,y_test_val,test_size = 0.5,random_state = 40)
```

```
X.shape
```

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,Sequential
tf.keras.__version__
```

```
model = Sequential ([
    layers.Dense(5, activation ='relu', input_dim= 4),
    # layers.Dense(30, activation ='relu'),
    layers.Dense(1, activation ='sigmoid'),    
])
```

```
# model.compile (optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy']) #0,8467
model.compile (optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy']) #0,8336
```

```
model.fit(x_train.values, y_train.values, epochs = 10, batch_size = 100)
```

```
results = model.evaluate(x_test.values, y_test.values)
results
```

```
y_pred_test = model.predict(x_test.values)

y_pred_test = (y_pred_test>0.5)
y_pred_test
```

```
y_test.values
```

```
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

confusion_matrix(y_test.values, y_pred_test)
```

```
X_top9 = X_norm[['RainToday', 'Humidity3pm', 'Rainfall', 'Humidity9am', 'WindGustSpeed', 'Temp3pm', 'Pressure9am', 'MaxTemp', 'Pressure3pm']]
X_top9.head()
x_train,x_test_val,y_train,y_test_val = train_test_split(X_top9,Y,test_size = 0.4,random_state = 40)
x_test, x_val, y_test, y_val = train_test_split(x_test_val,y_test_val,test_size = 0.5,random_state = 40)
model_top9 = Sequential( [
layers.Dense(10, activation='relu', input_dim=9),
layers.Dense(10, activation='relu'),
layers.Dense(1,activation='sigmoid')] )
model_top9.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model_top9.fit( x_train.values, y_train.values, epochs=10, batch_size=100 )
results = model_top9.evaluate(x_test.values, y_test.values)
results
```
