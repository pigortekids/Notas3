```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
import graphviz
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
caminho = "F:/gajo(a)s do quinto/iris.data.csv"
iris_dataset = pd.read_csv(caminho)
```

```python
sns.pairplot(iris_dataset, hue="rotulo")
```

```python
iris_dataset_num = iris_dataset.replace( { 'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2 } )
Y = iris_dataset_num['rotulo']
```

```python
colunas = iris_dataset_num.columns[:-1]
X = iris_dataset_num[colunas]
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split( X, Y, test_size=0.3, random_state=23 )
```

```python
classificador = DecisionTreeClassifier()
```

```python
classificador.fit(x_treino, y_treino)
```

```python
score_treino = classificador.score(x_treino, y_treino)
score_treino
```

```python
score_teste = classificador.score(x_teste, y_teste)
score_teste
```

```python
dot_graph = export_graphviz(classificador, out_file='iris.dot')
with open("iris.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph).view()
```
